```python
import cv2
import numpy as np

# Initialize video capture (replace with your video path or use 0 for webcam)
cap = cv2.VideoCapture('C:/Users/Mr.kishore/Downloads/2637-161442811_small.mp4')  # Replace 'video.mp4' with your video file path or 0 for webcam

# Create background subtractor using MOG2 (you can also try cv2.createBackgroundSubtractorKNN())
bg_subtractor = cv2.createBackgroundSubtractorMOG2(detectShadows=True)

# Set a threshold for scene change detection
scene_change_threshold = 2000  # This threshold determines how large the change must be to be considered a scene change

while True:
    ret, frame = cap.read()
    if not ret:
        break
    
    # Apply the background subtractor to the frame
    fg_mask = bg_subtractor.apply(frame)

    # Find contours of the foreground objects in the mask
    contours, _ = cv2.findContours(fg_mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    # Calculate the total area of the detected contours
    total_area = sum(cv2.contourArea(contour) for contour in contours)

    # Check if the total area exceeds the threshold for scene change detection
    if total_area > scene_change_threshold:
        print("Scene change detected!")

    # Display the foreground mask and original frame
    cv2.imshow('Foreground Mask', fg_mask)
    cv2.imshow('Original Frame', frame)

    # Break the loop if the user presses 'q'
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release the video capture and close all windows
cap.release()
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np

# Open the video capture (replace 'video_path' with your video file or use 0 for webcam)
cap = cv2.VideoCapture('C:/Users/Mr.kishore/Downloads/2637-161442811_small.mp4')

# Initialize the background model (using the first frame as the background)
ret, frame = cap.read()
if not ret:
    print("Failed to read video.")
    exit()

# Convert the first frame to grayscale and float32 for precision
background = np.float32(cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY))

# Set the learning rate for averaging (higher value means faster adaptation to new background)
learning_rate = 0.01

while True:
    ret, frame = cap.read()
    if not ret:
        break
    
    # Convert the current frame to grayscale
    gray_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    # Update the background model with the current frame using the learning rate
    cv2.accumulateWeighted(gray_frame, background, learning_rate)

    # Subtract the current frame from the background to detect moving objects
    foreground_mask = cv2.absdiff(gray_frame, cv2.convertScaleAbs(background))

    # Threshold the difference to create a binary mask of the foreground
    _, binary_mask = cv2.threshold(foreground_mask, 50, 255, cv2.THRESH_BINARY)

    # Display the results
    cv2.imshow('Foreground Mask', binary_mask)
    cv2.imshow('Original Frame', frame)

    # Break the loop if the user presses 'q'
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release the video capture and close all windows
cap.release()
cv2.destroyAllWindows()
```
