```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image
image = cv2.imread("property3.jpeg", cv2.IMREAD_GRAYSCALE)

# Define Prewitt Operator Kernels
prewitt_x = np.array([[1, 0, -1], [1, 0, -1], [1, 0, -1]])
prewitt_y = np.array([[1, 1, 1], [0, 0, 0], [-1, -1, -1]])

# Define Roberts Operator Kernels
roberts_x = np.array([[1, 0], [0, -1]])
roberts_y = np.array([[0, 1], [-1, 0]])

# Apply filters using cv2.filter2D
prewitt_x_edge = cv2.filter2D(image, -1, prewitt_x)
prewitt_y_edge = cv2.filter2D(image, -1, prewitt_y)
prewitt_edges = prewitt_x_edge + prewitt_y_edge

roberts_x_edge = cv2.filter2D(image, -1, roberts_x)
roberts_y_edge = cv2.filter2D(image, -1, roberts_y)
roberts_edges = roberts_x_edge + roberts_y_edge

# Display Results
plt.figure(figsize=(10, 5))
plt.subplot(1, 3, 1)
plt.imshow(image, cmap="gray")
plt.title("Original Image")
plt.axis('off')
plt.subplot(1, 3, 2)
plt.imshow(prewitt_edges, cmap="gray")
plt.title("Prewitt Edge Detection")
plt.axis('off')
plt.subplot(1, 3, 3)
plt.imshow(roberts_edges, cmap="gray")
plt.title("Roberts Edge Detection")
plt.axis('off')
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image
image = cv2.imread("property3.jpeg", cv2.IMREAD_GRAYSCALE)

### ** Laplacian of Gaussian (LoG) for Blob Detection**
# Apply Gaussian Blur
blurred = cv2.GaussianBlur(image, (5, 5), 0)
# Apply Laplacian
log = cv2.Laplacian(blurred, cv2.CV_64F)
# Convert to absolute values
log = np.uint8(np.absolute(log))

### ** Difference of Gaussian (DoG) for Blob Detection**
# Apply Gaussian Blur with two different kernel sizes
gaussian1 = cv2.GaussianBlur(image, (5, 5), 1)
gaussian2 = cv2.GaussianBlur(image, (9, 9), 2)
# Subtract the two blurred images
dog = cv2.absdiff(gaussian1, gaussian2)

# Display Results
plt.figure(figsize=(12, 5))

plt.subplot(1, 3, 1)
plt.imshow(image, cmap="gray")
plt.title("Original Image")
plt.axis('off')
plt.subplot(1, 3, 2)
plt.imshow(log, cmap="gray")
plt.title("Laplacian of Gaussian (LoG)")
plt.axis('off')
plt.subplot(1, 3, 3)
plt.imshow(dog, cmap="gray")
plt.title("Difference of Gaussian (DoG)")
plt.axis('off')
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image in grayscale
image = cv2.imread("property3.jpeg", cv2.IMREAD_GRAYSCALE)

### ** Laplacian of Gaussian (LoG) for Blob Detection**
# Apply Gaussian Blur
blurred_log = cv2.GaussianBlur(image, (5, 5), 0)
# Apply Laplacian to detect blobs
log = cv2.Laplacian(blurred_log, cv2.CV_64F)
log = np.uint8(np.absolute(log))  # Convert to unsigned 8-bit image

### ** Difference of Gaussian (DoG) for Blob Detection**
# Apply Gaussian Blur with different kernel sizes
gaussian1 = cv2.GaussianBlur(image, (3, 3), 0)
gaussian2 = cv2.GaussianBlur(image, (7, 7), 0)
# Subtract the two blurred images
dog = cv2.absdiff(gaussian1, gaussian2)

### ** Apply Thresholding to Highlight Blobs**
_, log_thresh = cv2.threshold(log, 30, 255, cv2.THRESH_BINARY)
_, dog_thresh = cv2.threshold(dog, 15, 255, cv2.THRESH_BINARY)

# Display Results
plt.figure(figsize=(12, 6))

plt.subplot(2, 3, 1)
plt.imshow(image, cmap="gray")
plt.title("Original Image")
plt.axis('off')
plt.subplot(2, 3, 2)
plt.imshow(log, cmap="gray")
plt.title("Laplacian of Gaussian (LoG)")
plt.axis('off')
plt.subplot(2, 3, 3)
plt.imshow(log_thresh, cmap="gray")
plt.title("LoG with Threshold")
plt.axis('off')
plt.subplot(2, 3, 4)
plt.imshow(dog, cmap="gray")
plt.title("Difference of Gaussian (DoG)")
plt.axis('off')
plt.subplot(2, 3, 5)
plt.imshow(dog_thresh, cmap="gray")
plt.title("DoG with Threshold")
plt.axis('off')
plt.tight_layout()
plt.show()
```
